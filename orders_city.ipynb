```python
#convert  Quantity Ordered to int
alldata= alldata.astype({'Quantity Ordered':'int'})
alldata= alldata.astype({'Price Each':'float'})
```

```python
# best-selling product
alldata.groupby('Product')['Quantity Ordered'].sum().plot(kind='bar')
```

```python
products=alldata.groupby('Product')['Quantity Ordered'].sum().index
quantitys=alldata.groupby('Product')['Quantity Ordered'].sum()
prices=alldata.groupby('Product')['Price Each'].mean()
```

```python
plt.figure(figsize=(40,24))
fig,axe1=plt.subplots()
axe2=axe1.twinx()
axe1.bar(products,quantitys,color='g')
axe2.plot(products,prices,'b-')
axe1.set_xticklabels(products,rotation='vertical',size=8)
```
